```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # note: you cannot submit to autolab with matplotlib included

def read_csv(file_path):
    return pd.read_csv(file_path)


def main():
    # Read the CSV file into a Pandas Dataframe
    amd_df = read_csv("stock-prices/AMD.csv")
    nvda_df = read_csv("stock-prices/NVIDA.csv")

    amd_reverse = amd_df.iloc[::-1]
    nvda_reverse = nvda_df.iloc[::-1]
    
    plt.style.use('ggplot')                  # Configure matplotlib to display like ggplot (R package)
    plt.rcParams['figure.figsize'] = (20, 6)  # Configure matplotlib figure size (inches, inches)
    amd_reverse.plot(kind='line', x='date', y='close', title="AMD stock prices")
    plt.show()
    
    plt.style.use('ggplot')                  # Configure matplotlib to display like ggplot (R package)
    plt.rcParams['figure.figsize'] = (20, 6)  # Configure matplotlib figure size (inches, inches)
    nvda_reverse.plot(kind='line', x='date', y='close', title="NVIDIA stock prices")
    plt.show()


if __name__ == "__main__":
    main()
```
