```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # note: you cannot submit to autolab with matplotlib included

"""
This lab will introduce you to the popular Pandas library and the pandas. Series (1D) and pandas.DataFrame (2D)
objects. These objects (dataframe especially) provides easy to use methods that make data cleaning and analysis
simple compared to other libraries.

Pandas is built on NumPy and is highly regarded for its easy-to-use data manipulation features.

We will be using weather data from the Chinese provincial-city, Chengdu - the home of giant pandas.
"""


def read_csv(file_path):
    return pd.read_csv(file_path)


def main():
    # Read the CSV file into a Pandas Dataframe
    amd_df = read_csv("stock-prices/AMD.csv")
    nvda_df = read_csv("stock-prices/NVIDA.csv")


    plt.style.use('ggplot')                  # Configure matplotlib to display like ggplot (R package)
    plt.rcParams['figure.figsize'] = (20, 6)  # Configure matplotlib figure size (inches, inches)
    amd_df.plot(kind='line', x='date', y='close', title="AMD stock prices")
    plt.show()
    
    plt.style.use('ggplot')                  # Configure matplotlib to display like ggplot (R package)
    plt.rcParams['figure.figsize'] = (20, 6)  # Configure matplotlib figure size (inches, inches)
    nvda_df.plot(kind='line', x='date', y='close', title="AMD stock prices")
    plt.show()

# Editing this section may cause your code to fail in Autolab
if __name__ == "__main__":
    main()
```
