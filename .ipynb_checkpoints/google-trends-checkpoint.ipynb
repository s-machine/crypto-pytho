```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

# get Google search trends via a wrapped python API
def search_trends(key_word):  
    pytrends = TrendReq(hl='en-US', tz=360)

    kw_list = [key_word]

    pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')

    result = pytrends.interest_over_time()
    return result

# plot the search trends 
def plot_5_years_trends(keyword):
    bitcoint_df = search_trends(keyword)
        
    plt.style.use('ggplot')                  
    plt.rcParams['figure.figsize'] = (20, 6)  
    bitcoint_df.plot(kind='line', y=keyword, title=(keyword + 'bitcoint search trends'))
    plt.show()

def main():
    plot_5_years_trends('bitcoin')
    plot_5_years_trends('blockchain')
    

if __name__ == "__main__":
    main()
```
