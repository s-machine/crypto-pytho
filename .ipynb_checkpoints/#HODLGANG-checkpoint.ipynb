# Google Trend Data as a Proxy for Public Opinion
After much trial and more error on scraping Twitter data, we moved on to gathering data on volume of Google Searches to gauge public interest in topics and terms related to crypto. In this notebook, we were particularly interested in the crypto slang term, "hodl." The term caught on after an early adopter drunkely posted in an online forum "HODL" when the price of BTC was tanking, meaning to say "HOLD." It is a unique term as it is only spoekn by true crypto/blockchain fanatics, but as the world is becoming more aware of digital assets and blockchain technologies they are learning and reasearching their term. 

The two datasets we used are 1) "HODL" searches since 11/2016 from Google Trends (CSV download), and 2) BTC price data since 11/2016.

```python
import os
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from re import sub
from decimal import Decimal
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')
```

```python
# import HODL data from Google Trends download
hodl = pd.read_csv("hodl_google_trends_data.csv")
```

```python
# build simple plot of HODL searches since 11/2016
weeks_hodl = pd.to_datetime(hodl['Week'],infer_datetime_format=True)
plt.figure(figsize=(15,8))
plt.plot_date(weeks_hodl, hodl['hodl'], '-', color="c")
plt.xlabel("Week")
plt.ylabel("Interest over time")
plt.title("HODL Searches Over Time")
```

```python
# import BTC price data from Coindesk download
btc_price = pd.read_csv("coindesk_bitcoin_price.csv")
```

```python
# BTC Price data is daily whereas Google Trends HODL data is weekly, need to clean up before merging
btc_week_clean = []
for date in btc_price['Date']:
    btc_week_clean.append(date[:-6])
btc_price['Date'] = btc_week_clean
```

```python
weeks_btc = pd.to_datetime(btc_price['Date'],infer_datetime_format=True)

merged_df = pd.merge(hodl, btc_price, how='left', left_on = 'Week', right_on = 'Date')
merged_df.drop(columns=['Date'], axis=1, inplace=True)
```

```python
# use seaborn to build clean visualization of BTC Price (one y-axis) and HODL searches (other y-axis)
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

sns.set_style("white")

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(weeks_hodl,merged_df['hodl'], color="#C22345")
ax.yaxis.label.set_text("HODL Searches")
ax.yaxis.label.set_color('#C22345')
ax.xaxis.label.set_color('black')
ax.tick_params(axis='y', colors='#C22345')
ax2 = ax.twinx()
ax2.plot(weeks_hodl,merged_df['Close_Price'], color="#3949BC")
ax2.yaxis.label.set_text("Bitcoin Price")
ax2.yaxis.label.set_color('#3949BC')
ax2.tick_params(axis='y', colors='#3949BC')
sns.despine(ax=ax, right=True, left=True)
sns.despine(ax=ax2, left=True, right=False)
ax2.spines['right'].set_color('white')
plt.title("Highly Correlated Bitcoin Price and HODL Searches")
```

This was an unexpected result! While we did hypothesis a correlation between BTC price and HODL searches, we didn't anticipate this much overlap. In fact, if you look closely the price seems to be a bit of a leading indicator for HODL searches (unfortunately not the other way around). But what is happening recently? The last few weeks on the chart are showing a separation. 

You might be wondering why the HODL searches axis ranges from 0 to 100--from Google Trends site:
> Numbers represent search interest relative to the highest point on the chart for the given region and time. A value of 100 is the peak popularity for the term. A value of 50 means that the term is half as popular. A score of 0 means there was not enough data for this term.
