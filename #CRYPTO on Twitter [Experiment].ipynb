```python
import os
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from re import sub
from decimal import Decimal
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
twt = pd.read_csv("twitter.csv")
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
stopwords.update(('http','RT','https','follower'))

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=60, 
        scale=3,
        random_state=69 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(17, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(twt['text'])
```

While we did not include this in our final dataset or presentation, it is somewhat interesting that words like "Coal" and "weird" and "Samsung" appeared in the data. We must keep in mind that this is only about 9,000 tweets from a single day and we'd probably see fewer unique words like these if we had more days in the data.
